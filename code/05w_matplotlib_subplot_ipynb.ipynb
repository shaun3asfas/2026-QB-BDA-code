```python
import matplotlib.pyplot as plt
# 전체 제목 — 어느 위치에서 호출해도 무관 (Figure 레벨)
plt.suptitle('전체 제목', fontsize=14)
# subplot 1 — 생성 직후 바로 title 지정
plt.subplot(2, 3, 1)
plt.plot([1, 4, 9, 16])
plt.title('제곱수')
# subplot 5
plt.subplot(2, 3, 6)
plt.plot([1, 2, 3, 4])
plt.title('선형')
# subplot 2
plt.subplot(2, 3, 2)
plt.plot([6, 5, 2, 1])
plt.title('감소')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 3, figsize=(7, 4))
fig.set_size_inches(6, 3)
# ax[0, 1].plot([1, 4, 9, 16])
ax[0][1].plot([1, 4, 9, 16])
ax[1, 0].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])
# 전체 제목 — Figure 레벨
fig.suptitle('전체 제목', fontsize=14, fontweight='bold')
# 각 subplot 제목 — Axes 레벨
ax[0, 0].set_title('(0,0) 빈 플롯')
ax[0, 1].set_title('(0,1) 제곱수')
ax[0, 2].set_title('(0,2) 빈 플롯')
ax[1, 0].set_title('(1,0) 선형')
ax[1, 1].set_title('(1,1) 빈 플롯')
ax[1, 2].set_title('(1,2) 감소')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6, 4))
# ax1 = fig.add_subplot(231)
ax1 = fig.add_subplot(2, 3, 1)
ax1.plot([1, 4, 9, 16])
ax1.set_title('제곱수')

ax2 = fig.add_subplot(233)
ax2.plot([1, 2, 3, 4])
ax2.set_title('선형')

ax3 = fig.add_subplot(235)
ax3.plot([6, 5, 2, 1])
ax3.set_title('감소')

fig.suptitle('전체 제목', fontsize = 14)
plt.tight_layout()
plt.show()
```

#####Series의 정렬

```python
import numpy as np
import pandas as pd

s =['희빈', '수영', '현수', '지호', '지민']
p = pd.Series([99, 84, 92, 65, 78], index=s)
p
```

```python
p.name = '수학성적'
p.index.name = 'st_name'
p
```

```python
p.sort_values() #시리즈의 값을 오름차순
```

```python
p.sort_values(ascending=False) #시리즈의 값을 내림차순
```

```python
p.sort_index() #index 값을 오름차순 ascending=True 와 결과 같음
```

```python
p.sort_index(ascending=False) #index 값을 내림차순
```

```python
s = pd.Series([1, 3, 5, np.nan, 6, 8], name='first')
s.index = list('abcdef')
print(s)
```

```python
print(s['b'])        # 레이블로 참조
print(s[1])          # 정수 위치로 참조
print(s['b':'e'])    # 레이블 슬라이싱 (e포함)
print(s[1:4])        # 정수 슬라이싱 (4미포함)
print(s[['a', 'c']]) # 여러 요소 선택
```

```python
dates = pd.date_range('20250101', periods=6)
print(dates)
```

```python
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list('ABCD'))
print(df)
```

```python
print(df.head())     # 상위 5행
print(df.tail(3))    # 하위 3행
print(df.index)      # 인덱스 확인
print(df.columns)    # 컬럼명 확인
print(df.values)     # 값(numpy 배열) 확인
```

```python
print(df.describe()) # 기술 통계
print(df.T)          # 행과 열 바꿈
```

```python
print(df.sort_index(axis=1, ascending=False))
```

```python
print(df.sort_values(by='B', ascending=False))
```

```python
print(df['A'])          # 단일 열(시리즈)
print(df[['A', 'C']])   # 여러 열(데이터프레임)
print(df[0:3])          # 행 슬라이싱
```

```python
print(df[df['A'] > 0])
```

```python
print(df.loc[dates[0]])            # 첫 번째 날짜 행
print(df.loc[:, ['A', 'B']])       # 모든 행의 A, B열
print(df.loc['20250102':'20250104', ['A', 'B']])
```

```python
print(df.iloc[3])                  # 4번째 행
print(df.iloc[3:5, 0:2])           # 3~4행, 0~1열
print(df.iloc[[1, 2, 4], [0, 2]])  # 특정 위치 추출
```

```python
print(df.at[dates[0], 'A'])        # 레이블로 단일값 접근
print(df.iat[1, 1])                # 위치로 단일값 접근
```
