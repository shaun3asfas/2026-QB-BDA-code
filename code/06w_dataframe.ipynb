```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
display_side_by_side(df, df.cumsum(axis=1))
```

```python
display_side_by_side(df, df.cumsum())
```

#####함수 df.query(조건) #조건을 만족하는 행을 표시

```python
df = pd.DataFrame({'A': range(1, 6),
                   'B': range(6, -3, -2),
                   'C': range(10, 5, -1),
                   'D clmns': range(3, -2, -1),
                   'E': ['python', 'java', 'python', 'go', 'python']})
df
```

```python
df.query('A > B')
```

```python
df[df.A > df.B]
```

#####열 이름을 바로 기술 가능

```python
df.query('C == 10')
```

```python
df.query('A > 0')
```

```python
df.query('E == "python"')
```

```python
df.query("E != 'python'")
```

#####열명 중간에 빈 문자열이 있는 경우

```python
df.query('A == `D clmns`')
```

```python
df.query('A == "D clmns"')
```

```python
df[df.A == df["D clmns"]]
```

```python
df.query('A > 3 | B > B')
```

```python
df.query('A > 2 & C < B')
```

```python
df.query("E in ['python', 'go']")
```

#####df.query() 내부의 질의 문자열 내부에서 변수를 사용

```python
num = 3
df.query('B > @num')
```

```python
num = 3
df.query(f'B > {num}') #f 문자열 내부에서 변수 num은 {num} 사용가능
```

```python
s = "E.str.contains('py')" #열 명.str.contains('부분문자열') 사용
df.query(s)
```

```python
s = "E.str.contains('g')" #열 명.str.startswith('시작문자열') 사용
df.query(s)
```

```python
s = "E.str.contains('on')" #열 명.str.endswith('끝나는문자열') 사용
df.query(s)
```
